# Exploring carpet plots

**Carpet plots**, as they are described in [Power (2017)](https://www.sciencedirect.com/science/article/pii/S1053811916303871) are a great way to quickly get an overview over a functional MRI image. I first encountered such a carpet plot with `fmriprep`.

As a useful tool for quality control, I was curious if one could get even more out of this, by slightly adjusting the plots. This notebook describes different such approaches.

**Note**: I want to point out that the functional image that I used for the example was already preprocessed (realigned, temporal filtered and spatially smoothed). It is therefore a "clean" fMRI image, and perhaps not the best example to show how a carpet plot can point to quality issues.

## Import relevant modules

```python
from nilearn import plotting, image
%matplotlib inline
import numpy as np
import nibabel as nb
import matplotlib.pyplot as plt
```

## Load and prepare some data

```python
# Load functional image
img_func = nb.load('sub-01_task-memory_run-01.nii.gz')
data = img_func.get_data()
```

```python
# Load GM, WM and CSF mask
img_gm = nb.load('sub-01_seg_gm.nii.gz')
img_wm = nb.load('sub-01_seg_wm.nii.gz')
img_csf = nb.load('sub-01_seg_csf.nii.gz')
```

```python
# Resample masks to functional space
mask_gm = image.resample_to_img(img_gm, img_func, interpolation='nearest').get_data() >= 0.5
mask_wm = image.resample_to_img(img_wm, img_func, interpolation='nearest').get_data() >= 0.5
mask_csf = image.resample_to_img(img_csf, img_func, interpolation='nearest').get_data() >= 0.5
```

```python
# Load the data for each segment
data_gm = data[mask_gm]
data_wm = data[mask_wm]
data_csf = data[mask_csf]
```

```python
# Remove those voxels without any variation over time
data_gm = data_gm[data_gm.std(axis=-1)!=0]
data_wm = data_wm[data_wm.std(axis=-1)!=0]
data_csf = data_csf[data_csf.std(axis=-1)!=0]
```

## Create simple carpet plot (without any corrections)

This carpet plot shows the timecourse of all voxels in the GM (above red line), followed by WM (between red and blue line) and CSF (below blue line).

```python
# Create carpet
carpet = np.row_stack((data_gm,
                       data_wm,
                       data_csf))
```

```python
# Plot carpet plot
plt.figure(figsize=(12, 6))
plt.imshow(carpet, aspect='auto', cmap='gray')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='r')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='b');
```

As you can see, plotting the original data gives us just information about which voxels have high and which have low value. But in respect to the timecourse, we don't really get any information.

## Create simple carpet plot with zscored data

The solution to getting more information about the timecourse it to standardize the data over time. This can be done with `scipy`'s `zscore`.

```python
from scipy.stats import zscore
```

```python
# Z-standardizes original carpet data
zcarpet = np.nan_to_num(zscore(carpet, axis=-1))
```

```python
# Plot carpet plot
plt.figure(figsize=(12, 6))
plt.imshow(zcarpet, aspect='auto', cmap='gray')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='r')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='b');
```

## Investigating the carpet plot

Now here we can already see some kind of structure in the functional images. There seem to be some strong signal changes volume 40, 100, 145 and 200.

Let's try to plot the average signal timecourse.

```python
plt.figure(figsize=(12, 2))
plt.plot(zcarpet.mean(axis=0));
```

Hmm... That doesn't help much. But what if we plot the signal to the power of 5?

```python
fig1 = plt.figure(figsize=(12, 6))
plt.imshow(zcarpet, aspect='auto', cmap='gray')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='r')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='b')

fig2 = plt.figure(figsize=(12, 2))
plt.plot(zcarpet.mean(axis=0)**5)
ax = fig2.get_axes()[0]
ax.set_xlim([0, data.shape[-1]]);
```

Much better! But in what way does this correspond to for example framewise displacement (FA)?

```python
import pandas as pd
confounds = pd.read_table('sub-01_task-memory_run-01_confounds.tsv')
plt.figure(figsize=(12, 2))
plt.plot(confounds['FD']);
```

Interesting. So the anomalies in the carpet plot are not strongly correlating to FD. So it seems we're getting some additional information out of it.

Are there perhaps other ways to get even more information from the carpet plots?

## Offset carpet plot

Let's start with using the three color channels to plot the timecourse with some offsets:
 - **R**: original timecourse
 - **G**: timecourse with +1 offset
 - **B**: timecourse with -1 offset

```python
# First let's rescale timecourse values to be between 0 and 255
pos_carpet = zcarpet - zcarpet.min(axis=-1)[..., None]
pos_carpet = (pos_carpet / pos_carpet.max(axis=-1)[..., None]) * 255
```

```python
# Creating individual timecourses
offset = 1
t_R = np.copy(pos_carpet)
t_G = np.column_stack((np.zeros((len(pos_carpet), offset)), pos_carpet[:, :-offset]))
t_B = np.column_stack((pos_carpet[:, offset:], np.zeros((len(pos_carpet), offset))))
```

```python
# Creating RGB carpet plot
offset_carpet = np.stack((t_R, t_G, t_B), axis=-1).astype('int')
```

```python
fig1 = plt.figure(figsize=(12, 6))
plt.imshow(offset_carpet, aspect='auto')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='r')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='b')

fig2 = plt.figure(figsize=(12, 2))
plt.plot(offset_carpet.mean(axis=0)**5)
ax = fig2.get_axes()[0]
ax.set_xlim([0, data.shape[-1]]);
```

## Derivative carpet plot

Similar to the approach above, let's use the RGB channel to plot three information at once. But this time, let's plot the derivatives of the timecourse.
 - **R**: original timecourse
 - **G**: first derivative
 - **B**: second derivative

```python
# Creating individual timecourses
t_R = np.copy(zcarpet)
t_G = np.column_stack((np.diff(zcarpet), np.zeros((len(zcarpet), 1))))
t_B = np.column_stack((np.diff(np.diff(zcarpet)), np.zeros((len(zcarpet), 2))))

# Rescale values so that they are between 0 and 255
t_R -= t_R.min(axis=-1)[..., None]
t_G -= t_G.min(axis=-1)[..., None]
t_B -= t_B.min(axis=-1)[..., None]

t_R = (t_R / t_R.max(axis=-1)[..., None]) * 255
t_G = (t_G / t_G.max(axis=-1)[..., None]) * 255
t_B = (t_B / t_B.max(axis=-1)[..., None]) * 255
```

```python
# Creating RGB carpet plot
diff_carpet = np.stack((t_R, t_G, t_B), axis=-1).astype('int')
```

```python
fig1 = plt.figure(figsize=(12, 6))
plt.imshow(diff_carpet, aspect='auto')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='r')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='b')

fig2 = plt.figure(figsize=(12, 2))
plt.plot(diff_carpet.mean(axis=0)**5)
ax = fig2.get_axes()[0]
ax.set_xlim([0, data.shape[-1]]);
```

# Power spectrum carpet plot

Now, if we don't plot the timecourse itself, but the power spectrum for each voxel, we might get some information about the three different regions. The thought behind this is that the GM regions should have a different power spectrum than the WM or the CSF region.

```python
from scipy.signal import welch
```

```python
# Compute power spectral density
TR = img_func.header['pixdim'][4]
freq, power_carpet = welch(zcarpet, fs=TR)
```

```python
fig1 = plt.figure(figsize=(12, 6))
plt.imshow(power_carpet, aspect='auto')
plt.hlines(data_gm.shape[0], 0, power_carpet.shape[1] - 1, colors='r')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, power_carpet.shape[1] - 1, colors='b')

fig2 = plt.figure(figsize=(12, 2))
border1 = len(data_gm)
border2 = len(data_gm) + len(data_wm)
plt.plot(power_carpet[:border1,:].mean(axis=0))
plt.plot(power_carpet[border1:border2,:].mean(axis=0))
plt.plot(power_carpet[border2:,:].mean(axis=0))
ax = fig2.get_axes()[0]
ax.set_xlim([0, power_carpet.shape[-1]])
ax.set_xticklabels(np.round(ax.get_xticks() / ax.get_xticks().max() * (TR / 2.), 3));
```

This works almost. Looking at the different regions, and the average power spectrum curve below the carpet plot, we can see that the signal in the CSF is different than the signal in GM or WM.

## PCA carpet plot

Another possible approach to explore the information that is stored in the different regions is PCA. If we compute a PCA on the GM data and than multiply the first three components (because of RGB) with all voxels, we should hopefully see a difference between the regions.

The thinking is as follows. The first three components explain the most variance in the data. Hopefully the components in the GM regions is fundamentally different from the other regions, so that we should see a different color patter in the individual regions.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
fit = pca.fit(zcarpet[:data_gm.shape[0] - 1, :])
rgb_component = fit.components_[:3, :]
```

```python
# Multiply the first three component with the carpet plot
t_R = np.prod((zcarpet, rgb_component[0, :]))
t_G = np.prod((zcarpet, rgb_component[1, :]))
t_B = np.prod((zcarpet, rgb_component[2, :]))

# Rescale values so that they are between 0 and 255
t_R -= t_R.min(axis=-1)[..., None]
t_G -= t_G.min(axis=-1)[..., None]
t_B -= t_B.min(axis=-1)[..., None]

t_R = (t_R / t_R.max(axis=-1)[..., None]) * 255
t_G = (t_G / t_G.max(axis=-1)[..., None]) * 255
t_B = (t_B / t_B.max(axis=-1)[..., None]) * 255
```

```python
# Creating RGB pca carpet plot
pca_carpet = np.stack((t_R, t_G, t_B), axis=-1).astype('int')
```

```python
fig1 = plt.figure(figsize=(12, 6))
plt.imshow(pca_carpet, aspect='auto')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='k', linestyles='dashed')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='k', linestyles='dashed')

fig2 = plt.figure(figsize=(12, 2))
plt.plot(zcarpet.mean(axis=0)**5)
ax = fig2.get_axes()[0]
ax.set_xlim([0, data.shape[-1]])

fig3 = plt.figure(figsize=(12, 2))
plt.plot(rgb_component.T)
ax = fig3.get_axes()[0]
ax.set_xlim([0, data.shape[-1]]);
```

This kind of worked. The three regions seem to have a bit different color tone. This becomes even clearer when we plot the average time course (in RGB) per region.

```python
# Get average RGB timecourse per region
gm_mean = pca_carpet[:data_gm.shape[0] - 1,...].mean(axis=0)
wm_mean = pca_carpet[data_gm.shape[0]:data_gm.shape[0] + data_wm.shape[0] - 1,...].mean(axis=0)
csf_mean = pca_carpet[data_gm.shape[0] + data_wm.shape[0]:,...].mean(axis=0)
```

```python
fig1 = plt.figure(figsize=(12, 6))
summary = np.stack((gm_mean, wm_mean, csf_mean), axis=0).astype('int')
plt.imshow(summary, aspect='auto')
```

## ICA carpet plot

How about the exact same thing as above, but this time using an ICA!

```python
from sklearn.decomposition import FastICA
ica = FastICA(n_components=100)
fit = ica.fit(zcarpet[:data_gm.shape[0] - 1, :])
rgb_component = fit.components_[:3, :]
```

```python
# Multiply the first three component with the carpet plot
t_R = np.prod((zcarpet, rgb_component[0, :]))
t_G = np.prod((zcarpet, rgb_component[1, :]))
t_B = np.prod((zcarpet, rgb_component[2, :]))

# Rescale values so that they are between 0 and 255
t_R -= t_R.min(axis=-1)[..., None]
t_G -= t_G.min(axis=-1)[..., None]
t_B -= t_B.min(axis=-1)[..., None]

t_R = (t_R / t_R.max(axis=-1)[..., None]) * 255
t_G = (t_G / t_G.max(axis=-1)[..., None]) * 255
t_B = (t_B / t_B.max(axis=-1)[..., None]) * 255
```

```python
# Creating RGB ica carpet plot
ica_carpet = np.stack((t_R, t_G, t_B), axis=-1).astype('int')
```

```python
fig1 = plt.figure(figsize=(12, 6))
plt.imshow(ica_carpet, aspect='auto')
plt.hlines(data_gm.shape[0], 0, data_gm.shape[1] - 1, colors='k', linestyles='dashed')
plt.hlines(data_gm.shape[0] + data_wm.shape[0], 0, data_gm.shape[1] - 1, colors='k', linestyles='dashed')

fig2 = plt.figure(figsize=(12, 2))
plt.plot(zcarpet.mean(axis=0)**5)
ax = fig2.get_axes()[0]
ax.set_xlim([0, data.shape[-1]])

fig3 = plt.figure(figsize=(12, 2))
plt.plot(rgb_component.T)
ax = fig3.get_axes()[0]
ax.set_xlim([0, data.shape[-1]]);
```

And how does the average RGB timecourse per region look like?

```python
# Get average RGB timecourse per region
gm_mean = ica_carpet[:data_gm.shape[0] - 1,...].mean(axis=0)
wm_mean = ica_carpet[data_gm.shape[0]:data_gm.shape[0] + data_wm.shape[0] - 1,...].mean(axis=0)
csf_mean = ica_carpet[data_gm.shape[0] + data_wm.shape[0]:,...].mean(axis=0)
```

```python
fig1 = plt.figure(figsize=(12, 6))
summary = np.stack((gm_mean, wm_mean, csf_mean), axis=0).astype('int')
plt.imshow(summary, aspect='auto')
```
